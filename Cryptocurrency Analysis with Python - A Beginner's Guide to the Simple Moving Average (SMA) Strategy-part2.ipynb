## The SMA crossover strategy

<p style="text-align: justify;"> Technical analysis (TA) is a methodology used mostly in the financial market to forecast the direction of prices for various instruments, including cryptocurrencies [1]. TA is based on the idea that identification of previous price changes provides a relative accurate prediction of future price trajectories. TA uses different mathematical indicators that generate signals to buy or to sell financial  securities, indicating potential trend reversals. One frequently used indicator in TA is the Simple Moving Average (SMA), which calculates the average of closing prices over a specific period within a given range (as discussed in Part 1 of this tutorial).  A popular practice involves combining two SMAs, a short-term one like SMA50 and a long-term one like SMA200, each covering different time frames. This combination results in the well-known Simple Moving Average (SMA) Crossover Strategy.  The SMA Crossover Strategy is a powerful tool for monitoring financial markets and has been adapted for various applications, including tracking COVID-19 cases [1]. The primary objective of this strategy is to identify potential trend changes and generate signals (buy or sell) based on the crossovers of these moving averages. 
    

**The two key signals in the SMA crossover strategy are the "Golden Cross" and the "Death Cross."**  

**Golden Cross:**  
<p style="text-align: justify;">The Golden Cross is a significant technical event that takes place when the shorter-term SMA [SMA(50)] crosses above the longer-term SMA [SMA(200)]. This occurrence is considered a bullish signal (see Figure below) and indicates a potential upward trend reversal or continuation. When traders observe the Golden Cross, they often interpret it as a signal to go long (meaning to buy) or maintain a long position, anticipating further price appreciation.</div>



**Death Cross:**
<p style="text-align: justify;">The Death Cross occurs when the shorter-term SMA [SMA(50)] crosses below the longer-term SMA [SMA(200)]. It is considered a bearish signal (see Figure below) and indicates a potential downward trend reversal or continuation. When the Death Cross is observed, traders may interpret it as a signal to go short (meaning to sell) or hold a short position, anticipating further price depreciation.</div>  

<p style="text-align: justify;">It's important to note that the SMA crossover strategy is a trend-following strategy. It aims to capture trends and is most effective when the market is trending. In choppy or sideways markets, the strategy may generate false signals, leading to potentially unprofitable trades.</div>



### Risk Management:
<p style="text-align: justify;">Incorporating effective risk management practices is of utmost importance when employing the SMA crossover strategy or any trading approach. Traders should prioritize implementing stop-loss orders to define predetermined exit points and limit potential losses if the trade moves against them. Furthermore, determining appropriate position sizes based on their risk tolerance and overall trading plan is vital to safeguard their capital.</div>


### False Signals and Whipsaws:
<p style="text-align: justify;">It is essential to remain cautious about potential false signals generated by SMA crossovers, especially in volatile or choppy markets. Traders may encounter situations where the price crosses the moving averages multiple times without establishing a clear trend. These whipsaw movements can result in losses or reduced profitability.</div>


### Confirmation and Additional Analysis:
<p style="text-align: justify;">While SMA crossovers can serve as valuable signals, traders often seek confirmation by employing additional technical indicators or chart patterns. Popular complementary indicators include the Relative Strength Index (RSI), Moving Average Convergence Divergence (MACD), and identifying support and resistance levels (to be discussed in future tutorials).</div>


<p style="text-align: justify;">As with any trading strategy, it is essential to <b>backtest (see below)</b> the SMA crossover strategy on historical data and conduct thorough research before implementing it in live trading. Traders should also adapt the strategy to suit their individual trading styles and preferences [2].</div>

The code below has been generated in the first part of this [tutorial](https://medium.com/coinmonks/cryptocurrency-analysis-with-python-a-beginners-guide-to-the-simple-moving-average-sma-557352a9f59b). Briefly, thecode retrieves historical price data for the cryptocurrency Ethereum (ETH) using the yfinance library. It then processes the data by calculating the (Simple) Returns and Log-Returns for ETH's price and subsequently calculates the 50-day and 200-day Simple Moving Averages (SMAs) for smoothing and trend identification. Finally, plots the price along with the long-term and short-term moving averages.

![sma1](SMApart1.jpg)


```python
import yfinance as yf
import mplfinance as mpf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define the cryptocurrency symbol and the date range
symbol = ['ETH-USD']  # Ethereum
start_date = '2018-01-01'
end_date = '2023-07-24'

# Retrieve the historical price data for the specified cryptocurrency symbols
data = yf.download(symbol, start=start_date, end=end_date)
df = data["Close"].to_frame() #Extracts only the "Close" column
df.rename(columns={"Close":"Price"}, inplace=True) # Rename column

# Calculate (Simple) Returns & Log-Returns
df["(Simple) Returns"]= df["Price"].pct_change()## Calculate (Simple) Returns
df["Log-Returns"]= np.log(df["Price"].div(df["Price"].shift()))## Calculate Log-Returns

# Create long-term moving and short-term moving average columns
short_period = 50
long_period=200
df["SMA (50)"] = df["Price"].rolling(short_period).mean()
df["SMA (200)"]= df["Price"].rolling(long_period).mean()

# Drop rows with missing values (NaN) resulted from calculating returns
df.dropna(inplace=True)

# Plot Price, long-term moving and short-term moving averages
df[["SMA (200)","SMA (50)", "Price"]].plot(title = F"ETH with SMA({short_period}) and SMA({long_period})" , figsize=(10,6))
# Set y-axis title
plt.ylabel("Price ($)")
df.head(3)
```

Next, we will implement a long-only strategy based on the SMA crossover. This approach entails holding an asset, such as Ethereum, when we anticipate prices will rise, and selling the asset when we believe prices will decline. To facilitate this strategy, we will create an additional `Position column` in the dataset (see code below). In the context of this strategy, a Position value of 1 will indicate a bullish signal when the shorter-term SMA (SMA(50)) crosses above the longer-term SMA (SMA(200)), prompting us to take a long position and hold the asset (Ethereum). On the other hand, when the shorter-term SMA crosses below the longer-term SMA, we will adopt a neutral stance, equivalent to a zero position. This neutral position implies that we will not be holding the asset, anticipating potential price decreases.By incorporating the Position column and interpreting its values based on SMA crossovers, we can effectively execute a long-only strategy, positioning ourselves to benefit from upward price trends while avoiding unnecessary exposure during potential downward movements.

In the code below, the condition `df["SMA (50)"] > df["SMA (200)"`] checks if the shorter-term SMA (SMA(50)) is greater than the longer-term SMA (SMA(200)) at each row. If the condition is true, meaning the SMA(50) crosses above SMA(200), the corresponding value in the Positions column will be set to 1, indicating a buy signal (bullish signal ). Otherwise, if the condition is false, the value will be set to 0, indicating a sell or neutral signal.

```python
# Create a Position column (1 for buy, -1 for sell)
df['Positions'] = np.where(df["SMA (50)"] > df["SMA (200)"], 1, 0)
df.loc[:,["SMA (50)", "SMA (200)", "Positions"]].plot(title = F"ETH with SMA({short_period}), SMA({long_period}) and Poistions" , figsize=(10,6), secondary_y="Positions")
df.head(3)
```

## Trading Costs

<p style="text-align: justify;">  From the Position column coded above (and the graph) we can deduce that for long periods we are going long or shorts having only few transactions (of selling or bying) with the relativce costs. Altough, to have a realistic approach we have to caluclate them. To calulate the number of trades (with their costs) we must capsure when we have a change from long to short position (lor form short to long positions). For this reason we cretae a new column in the DataFrame called 'Trades', which calculates the difference between consecutive rows in the 'Positions' column using the diff() function.This column represents the trading actions, where a value of 1 indicates a trade was made (buy or sell) and a value of 0 indicates no trade was made (neutral position).</div>

```python
df["Trades"]=df["Positions"].diff().fillna(0).abs()
print(df["Trades"].value_counts())
df.head(3)
```

<p style="text-align: justify;"> The cost of trading cryptocurrencies consists of several components, including commissions, spread, and slippage. These costs can significantly impact the overall profitability of any trading strategy and should be taken into account seriusly by traders.</div>

**Commissions:**  
<p style="text-align: justify;"> Commissions are fees charged by the exchange platform for executing a trade. When you buy or sell cryptocurrencies, the exchange typically charges a small percentage of the trade value as a commission. The commission fees can vary depending on the exchange, the trading volume, and the type of order (market order, limit order, etc.). Traders should be aware of the commission structure of the exchange they are using to ensure they understand the costs associated with their trades.</div>

**Spread:**  
<p style="text-align: justify;"> The spread refers to the difference between the bid price (the highest price a buyer is willing to pay) and the ask price (the lowest price a seller is willing to accept) for a particular cryptocurrency pair. It is essentially the cost of trading between the buying and selling prices. The spread can vary based on the liquidity of the cryptocurrency and the exchange's order book. Cryptocurrencies with high liquidity and trading volumes usually have tighter spreads, while less liquid cryptocurrencies may have wider spreads. As a result, traders will pay a higher cost when buying or selling at the market price due to the spread.</div>

**Slippage:**  
<p style="text-align: justify;"> Slippage occurs when the execution price of a trade differs from the expected price at the time the order was placed. This can happen during periods of high volatility or low liquidity. For example, if a trader places a market order to buy a cryptocurrency, the actual price at which the order is filled may be slightly higher than the displayed price due to a sudden increase in demand. Similarly, when selling, the actual price may be slightly lower than the displayed price due to a sudden decrease in demand. Slippage can result in additional costs and can affect the overall profitability of the trade.</div>

Together, **commissions, spread, and slippage** represent the total cost of trading cryptocurrencies. These costs can vary significantly between different exchanges and can impact the trading strategy and profitability of traders and investors. It is essential for traders to be aware of these costs and consider them in their trading plans and risk management strategies. Traders should also explore various exchanges to find the most cost-effective options for their specific trading needs.

The code below calculates the total costs using the formula `Costs = np.log(1 - commissions) + np.log(1 - spread_slippage)` by considering both the commission and the combined impact of spread and slippage. The part `1 - commissions` (same logic for the part 1 - spread_slippage), subtracts the commission cost from 1. The reason for subtracting the commission from 1 is to represent the remaining fraction of the trade value after deducting the commission. For example, if the commission is 0.00075 (0.075%), then 1 - 0.00075 would be 0.99925, which means the remaining trade value is 99.925% after deducting the commission (see below in the section `Backtesting our strategy`). The natural logarithm is used to convert these costs into a logarithmic scale, which helps in understanding and analyzing the impact of these costs on the overall profitability of the trading strategy (see below, section `Backtesting our strategy`).

```python
comissions = 0.00075
Spead_Slippage = 0.0001
Costs = np.log(1-comissions)+np.log(1-Spead_Slippage)
Costs
```

## Backtesting our strategy

<p style="text-align: justify;">In the first part of our tutorial (Part 1), we calculated the daily Log-Returns for the Buy and Hold Strategy, which involves acquiring a cryptocurrency like Ethereum and holding it for an extended period. This Buy and Hold Strategy serves as benchmark for comparing with the Simple Moving Average (SMA) Crossover Strategy explained in this part.To compare the two strategies, we need to calculate the Returns for our SMA Crossover Strategy. When the position is long (holding Ethereum) on a particular day, the Return of our strategy is equal to the Buy and Hold return. Conversely, whenever the position is short (selling Ethereum), the Return of our strategy is zero (indicating no gain or loss). However, it's crucial to pay attention because both prices and SMAs are based on closing prices at the end of each day. Therefore, if we want to calculate the return of our strategy for the next day, we must shift the position by one timestamp (or one trading day).


To derive the strategy returns, we multiply the position (shifted by one timestamp/trading day) with the Log-Returns column. This calculation allows us to account for the changes in position between consecutive trading days and obtain the strategy returns.

Moreover, we calculate the net Strategy Returns Net after accounting for the costs associated with each trade. The trading costs include commissions, spread, and slippage, which can impact the overall profitability of the strategy. By considering these costs, we obtain a more accurate representation of the strategy's performance in real-world trading conditions</div>.

```python
# Calculate Returns for our startegy
df["Strategy Returns"] = df["Positions"].shift(1) * df["Log-Returns"]
df["Strategy Returns Net"] = df["Strategy Returns"]+df["Trades"]*Costs

# Drop rows with missing values (NaN) resulted from calculating Strategy Returns
df.dropna(inplace=True)
df.head(3)
```

<p style="text-align: justify;">Since Log-Returns are additive, calculating the sum over the columns Log-Returns, Strategy Returns, and Strategy Returns Net allows us to compare the performance of the strategy relative to the Buy and Hold Strategy. By comparing the values below, we observe that the net return (Strategy Returns Net) is slightly lower than the original return (Strategy Returns) due to the deduction of trading costs. However, both (Strategy Returns Net & Strategy Returns) outperform the Buy and Hold Strategy, indicating that our strategy could be profitable. This comparison suggests that implementing the SMA crossover strategy has the potential to generate positive returns even after considering trading costs.</div>

```python
df[["Strategy Returns", "Log-Returns", "Strategy Returns Net"]].sum()
```

The code below calculates and stores the cumulative returns for the three strategies in the DataFrame df. These strategies are Log-Returns, representing raw returns, Strategy Returns, representing returns without considering costs, and Strategy Returns Net, representing returns after accounting for trading costs. 

The cumulative returns (normalized with base of 1) help us understand how the strategies have performed over time, and they provide insights into the overall profitability and growth of the investments.

```python
df["Cumulative Returns"] = df["Log-Returns"].cumsum().apply(np.exp)
df["Cumulative startegy Returns"] = df["Strategy Returns"].cumsum().apply(np.exp)
df["Cumulative startegy Returns Net"] = df["Strategy Returns Net"].cumsum().apply(np.exp)
df.tail(3)
```

```python
df[["Cumulative Returns","Cumulative startegy Returns", "Cumulative startegy Returns Net"]].plot(title = F"ETH" , figsize=(10,6))
```

The result of the code below shows the annualized returns for three different strategies: "Strategy Returns," "Log-Returns," and "Strategy Returns Net."  The higher the value, the greater the annualized return achieved by that strategy.

```python
Annualized_Returns = (df[["Strategy Returns", "Log-Returns", "Strategy Returns Net"]].mean()) * 365.25
Annualized_Returns
```

<p style="text-align: justify;">The result of the code below shows the annualized risk for each strategy, expressed as a standard deviation measure. The higher the value, the greater the risk or volatility associated with that strategy. Our strategy has a much higher risk/volatility compared to the Buy and Hold strategy.</div>

```python
Annualized_Risk= df[["Strategy Returns", "Log-Returns", "Strategy Returns Net"]].std() * np.sqrt(365.25)
Annualized_Risk
```

<p style="text-align: justify;"> Our SMA Crossover Strategy exhibits a greater reward (mean return) compared to the Buy and Hold Strategy. This means that, on average, our Strategy has generated higher returns over the analyzed period than simply holding the cryptocurrency (Buy and Hold Strategy).However, it's essential to consider that the SMA Crossover Strategy has a higher annualized risk than the Buy and Hold Strategy. This implies that our strategy experiences greater volatility or fluctuations in returns over time compared to the Buy and Hold Strategy. While higher returns are desirable, it's crucial to strike a balance between reward and risk. A higher-risk strategy may lead to potential larger gains, but it also carries the possibility of larger losses.</div>
